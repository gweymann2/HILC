```python
from mask_beam_hilc import *

channels = sorted(["L1-040", "L2-050", "L1-060", "L2-068", 'L3-068', "L1-078", 
                   "L3-089", "L4-100", "L4-140", "M1-100", "M2-119", "M2-166", 
                   "M1-195", "H2-235", "H1-280", "M1-140", "H2-337", "H3-402"],
                  key = lambda x: x[-3:])

fbins_sqrt = lambda a, x : int(np.sqrt(x)*a)
a = 162 ; lbins = np.array([fbins_sqrt(a, x) for x in range(1000) if fbins_sqrt(a, x) < 1536])

nsim = 1

output_dic = Cl_Wl(nsim = nsim, 
                   channels = channels, 
                   fg_models = ["pysm_dust_0"], 
                   make_noise = True, 
                   per_cent_to_keep_ = 85, 
                   apo_mask_deg_ = 2, 
                   smooth_mask_deg_ = 2, 
                   lbins = lbins, 
                   seed_cmb = np.random.randint(0,10000), 
                   do_map = True, 
                   verbose = True,)
```

```python
p_input = np.sqrt(output_dic['map_cmb'][1]**2+output_dic['map_cmb'][2]**2)
print('input p map') ; hp.mollview(p_input, unit='mK', norm='hist') ; plt.show()

p_output = np.sqrt((output_dic['map1sim_clean'][1])**2+(output_dic['map1sim_clean'][2])**2)
print('rebuilt p map') ; hp.mollview(p_output, unit='mK', norm='hist') ; plt.show()

p_output = np.sqrt((output_dic['map1sim_noise'][1])**2+(output_dic['map1sim_noise'][2])**2)
print('rebuilt noise p map') ; hp.mollview(p_output, unit='mK', norm='hist') ; plt.show()

ind = np.random.randint(0,len(p_output))
print(p_output[ind],p_input[ind])

resi_percent = 100*abs((p_output-p_input)/p_input)
#resi_percent = np.where(resi_percent<100, resi_percent, resi_percent*1e-10)
#resi_percent[(resi_percent > 100) | (resi_percent == np.nan)] = 100
print('input p map - rebuilt p map') ; hp.mollview((resi_percent), unit='%', norm='log') ; plt.show()
```
